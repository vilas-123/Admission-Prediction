```
# @title
import pandas as pd
import matplotlib.pyplot as plt

# Step 1 & 2: Filter data based on selected site and DIC
selected_site = 'VJW'
selected_dic = 'DIC Iron'

filtered_df = df[(df['site'] == selected_site) & (df['DIC'] == selected_dic)]

# Step 3: Find the top 10 unique "Classification" values with their "Incident Subclassification"
top_classifications = (
    filtered_df
    .groupby(["Classification", "Incident Sub Classification"])
    .size()
    .reset_index(name="Count")
    .sort_values(by="Count", ascending=False)
    .head(10)
)

# Step 4: Find the corresponding top 10 locations for the selected "Classification" and "Incident Subclassification"
top_locations = (
    filtered_df[
        filtered_df["Classification"].isin(top_classifications["Classification"])
    ]["Location"]
    .value_counts()
    .head(10)
)

# Step 5: Visualization
fig, axes = plt.subplots(2, 1, figsize=(12, 10))

# Plot the top classifications
axes[0].barh(
    top_classifications["Classification"] + " - " + top_classifications["Incident Sub Classification"],
    top_classifications["Count"],
)
axes[0].set_xlabel("Count")
axes[0].set_ylabel("Classification & Incident Sub Classification")
axes[0].set_title(f"Top 10 Classifications for Site: {selected_site}, DIC: {selected_dic}")

# Plot the top locations
axes[1].barh(top_locations.index, top_locations.values)
axes[1].set_xlabel("Count")
axes[1].set_ylabel("Location")
axes[1].set_title(f"Top 10 Locations for Site: {selected_site}, DIC: {selected_dic}")

# Adjust layout
plt.tight_layout()
plt.show()
```

```
import pandas as pd
df = pd.read_csv("/content/incident_data_apr22_sep23.csv")
df
```

```
df["Incident Date"] = pd.to_datetime(df["Incident Date"], errors="coerce")

df["Incident Month"] = df["Incident Date"].dt.to_period("M")

# Group by site and month to count incidents
site_monthly_trend = df.groupby(["site", "Incident Month"]).size().unstack(level=0).fillna(0)
```

```
# Select relevant columns
selected_columns = ["site", "Department", "Severity", "Incident Details", "DIC", "Location", "Classification"]
df_selected = df[selected_columns]

# Check for missing values and clean data
df_selected = df_selected.dropna(subset=["Incident Details"])  # Remove rows with missing text descriptions

# Fill missing categorical values with "Unknown"
df_selected.fillna("Unknown", inplace=True)

# Display the first few rows after reloading
df_selected.head()
```

```
import nltk
import shutil
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# Download necessary NLTK resources
nltk.download("punkt")
nltk.download("stopwords")

# Text preprocessing function
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    tokens = word_tokenize(text)  # Tokenize words
    tokens = [word for word in tokens if word.isalpha()]  # Remove punctuation/numbers
    stop_words = set(stopwords.words("english"))  # Load stop words
    tokens = [word for word in tokens if word not in stop_words]  # Remove stop words
    return " ".join(tokens)

# Apply preprocessing to the "Incident Details" column
df_selected["Processed_Incident"] = df_selected["Incident Details"].apply(preprocess_text)

# Convert processed text into numerical features using TF-IDF
tfidf_vectorizer = TfidfVectorizer(max_features=500)
incident_tfidf = tfidf_vectorizer.fit_transform(df_selected["Processed_Incident"])

# Use KMeans clustering to classify incidents into risk levels
kmeans = KMeans(n_clusters=3, random_state=42)  # 3 risk categories: Low, Medium, High
df_selected["Risk_Level"] = kmeans.fit_predict(incident_tfidf)

# Map risk levels to human-readable categories
risk_mapping = {0: "Low Risk", 1: "Medium Risk", 2: "High Risk"}
df_selected["Risk_Category"] = df_selected["Risk_Level"].map(risk_mapping)

# Display the first few rows to verify classification
df_selected[["site", "Department", "DIC", "Location", "Risk_Category"]].head()
```
